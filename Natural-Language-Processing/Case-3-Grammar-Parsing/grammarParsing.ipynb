## Grammar Parsing using NLTK

Grammar Parsing using Natural Language ToolKit  
Buatlah representasi grammar production dalam bentuk Context Free Grammar (CFG) 
beserta parsing tree-nya untuk menangani enam kalimat berikut: 
<br>
- The farmer loaded the cart with sand  
- The farmer loaded sand into the cart  
- The farmer filled the cart with sand  
- The farmer filled sand into the cart  
- The farmer dumped the cart with sand  
- The farmer dumped sand into the cart

## Import Library

```python
import nltk
from nltk import CFG
from nltk.parse import ChartParser
```

## Grammar Parsing

```python
# Definisikan grammar

grammar = CFG.fromstring("""
  S -> NP VP
  NP -> Det N | N
  VP -> V NP PP | V NP
  PP -> P NP
  Det -> 'The' | 'the' 
  N -> 'farmer' | 'cart' | 'sand'
  V -> 'loaded' | 'filled' | 'dumped'
  P -> 'with' | 'into'
""")

# Kalimat-kalimat yang akan dianalisis

sentences = [
    ['The', 'farmer', 'loaded', 'the', 'cart', 'with', 'sand'],
    ['The', 'farmer', 'loaded', 'sand', 'into', 'the', 'cart'],
    ['The', 'farmer', 'filled', 'the', 'cart', 'with', 'sand'],
    ['The', 'farmer', 'filled', 'sand', 'into', 'the', 'cart'],
    ['The', 'farmer', 'dumped', 'the', 'cart', 'with', 'sand'],
    ['The', 'farmer', 'dumped', 'sand', 'into', 'the', 'cart'],
]

# Parser untuk grammar

parser = nltk.ChartParser(grammar)

# Analisis setiap kalimat dan tampilkan parsing tree

for i, sentence in enumerate(sentences):

    print(f"Parsing tree {i+1} for: {' '.join(sentence)}")
    try:

        # demonstrate_nlp_parsing(sentence, grammar)

        # Menggunakan parser untuk menghasilkan pohon parsing

        for tree in parser.parse(sentence):

            print("\nBasic Tree: ")
            print(tree)

            print("\nTree Pretty Print:\n")
            tree.pretty_print()  # Menampilkan pohon parsing dalam format teks

            # Uncomment this to display the trees in GUI

            # tree.draw()

    except ValueError:
        print("Sentence cannot be parsed")
    print("\n" + "-"*50 + "\n")
```
